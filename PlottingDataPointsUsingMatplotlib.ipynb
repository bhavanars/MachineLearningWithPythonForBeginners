```python
#scattering data points using matplotlib
import matplotlib.pyplot as plt  #importing matplotlib library
import numpy as np               #importing numpy library
X = np.linspace(-2, 2, 101) 
Y = 2.2 * X + np.random.randn(*X.shape) * 0.4 + 0.3   # generating X and Y values using numpy linspace and random function
plt.figure() 
plt.scatter(X,Y)      #scattering points
plt.plot (X, 0.2+ 3 * X)  #plotting points using linear regression


```
