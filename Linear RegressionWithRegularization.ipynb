```python
from sklearn.datasets import load_boston  #importing dataset from sklearn packages
from matplotlib import pyplot as plt  #importing matplotlib for plotting data
import random
import math
boston = load_boston()
plt.scatter(boston.data[:,5] , boston.target , color = 'r')
w0 = random.randint(3,4)             #taking random values for w0 and w1
w1 = random.randint(6,8) 
w = [w0 , w1]
print(w)
x = boston.data[:,5]
y = boston.target
y_predicted = [ ]
residual = [ ]
e_weight = [ ]
e= 0
n = len(x)
lamda = 0.5
  #implementing linear regression function
for i in range(0,len(w)):
 e_weight = e_weight +  [e/2 + (lamda * w[i])]    
for i in x:
    z = e_weight[0] + e_weight[1] * i
    y_predicted = y_predicted + [z] 
for i in range(0,len(y)):
  residual = residual + [y[i] - y_predicted[i]]
for i in range(0 , len(residual)):
    e = e + math.pow(residual[i] , 2)
rmse = math.sqrt(e/n)


print(rmse) 
plt.plot(x , y_predicted)
           
```
