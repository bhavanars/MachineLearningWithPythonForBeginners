```python
from sklearn.datasets import load_boston  #importing dataset from sklearn packages
from matplotlib import pyplot as plt  #importing matplotlib for plotting data
import random
import math
boston = load_boston()
plt.scatter(boston.data[:,5] , boston.target , color = 'r')
w0 = random.randint(1,2)             #taking random values for w0 and w1
w1 = random.randint(2,4)             
x = boston.data[:,5]
y = boston.target
y_predicted = [ ]
residual = [ ]
e= 0
n = len(x)
for i in x:
    z = w0 + w1 * i
    y_predicted = y_predicted + [z]   #implementing linear regression function
plt.plot(x , y_predicted)
for i in range(0,len(y)):
  residual = residual + [y[i] - y_predicted[i]]
for i in range(0 , len(residual)):
    e = e + math.pow(residual[i] , 2)
rmse = math.sqrt(e/n)
print(rmse)      
           
```
